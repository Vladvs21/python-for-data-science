# IІ. Data Visualization.

2. (10б) Створіть pie chart, який би показував співвідношення виживших до загиблих.

```python
alive_num = df['Survived'].sum()
dead_num = len(df) - alive_num

labels = ['Alive', 'Dead']
data = [alive_num, dead_num]
colors = ['green', 'red']
explode = (0.1, 0)

plt.pie(data, explode=explode, labels=labels, colors=colors, autopct='%1.1f%%', shadow=True, startangle=140)
plt.title('Ratio of Alive to Dead')
plt.show()
```

За візуалізацією можна побачити, що всього трохи більше третини (а саме 38.4%) пасажирів вижило, в той час як загинуло аж 61.6%.

3. (20б) Створіть bar chart, який би показував співвідношення загиблих до виживших для кожної статі. Для groupby використовуйте атрибут as_index=False.


image.png

```python
grouped_data = df.groupby(['Survived', 'Sex'], as_index=False)['PassengerId'].count()
pivot_data = grouped_data.pivot(index='Sex', columns='Survived', values='PassengerId')

colors = ['tab:blue', 'tab:orange']

pivot_data.plot(kind='bar', color=colors, figsize=(6, 4), width=0.8)
plt.title('Survived vs Dead by Sex')
plt.xlabel('Sex')
plt.ylabel('count')
plt.xticks(rotation=0)
plt.legend(title='Survived')
plt.show()
```

Загалом жінок було набагато менше. При цьому жінок вижило набагато (+- в 2 рази) більше в абсолюті, ніж чоловіків

4. (20б) Відобразіть кількість пропущених значень в датасеті по кожній із змінних.

image.png

```python
missing_values = df.isnull().sum().sort_values(ascending=False)

plt.figure(figsize=(6, 5))
missing_values.plot(kind='bar', color='tab:blue')
plt.title('The amount of nullable values per column')
plt.xticks(rotation=60, ha='right')
plt.show()
```

Найбільше пропущених значенб в колонках Cabin та Age. Незначні пропуски має стовбець Embarked, в той час як інші заповнені повністю

5. (10б) Побудуйте наступний графік, дайте йому назву та опишіть, що власне на них відображається.

```python
import seaborn as sns
import matplotlib.pyplot as plt

df_for_violinplot = df.dropna(subset=['Age'])

plt.figure(figsize=(6, 4))
sns.violinplot(data=df_for_violinplot, x='Pclass', y='Age', hue='Survived', split=True)
plt.xlabel('Pclass')
plt.ylabel('Age')
plt.legend(title='Survived')
plt.show()
```

Видно, що в 1 класі були більш дорослі пассажири, при цьому тут і найбільший рівень виживаємості. Найбільший рівень смертності в 2 класі, а наймолодший найбільший показник смертності в 3 класі

6. (13б) Побудуйте графік розподілу частот по віку пасажирів.


image.png

```python
age_data = df['Age'].dropna()

plt.figure(figsize=(6, 4))
plt.hist(age_data, bins=20, color='tab:blue', edgecolor='black')
plt.title("Passengers' ages distribution")
plt.xlabel('Age')
plt.ylabel('Count')
plt.grid(False)

median_age = np.median(age_data)
mean_age = np.mean(age_data)
plt.axvline(x=median_age, color='red', linewidth=1.5, label='median')
plt.axvline(x=mean_age, color='green', linewidth=1.5, label='mean')

plt.legend()
plt.show()
```

Тут можна побачити розподіл вікових категорій пассажирів. Переважна більшість мала вік від 16 до 40 років. При цьому найбільше було людей з віком від 25 до 30. Середній вік пасажирів склав близько 30 років. А медіана - 28.

Чи існує якась кореляція між віком, вартістю квитка за виживаємістю

```python
numerical_variables = ['Survived', 'Age', 'Fare']

# Creating a correlation matrix
correlation_matrix = df[numerical_variables].corr()

# Plotting the heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap between Fare, Age and Survival rate')
plt.show()
```

За результатами можна зробити висновок, що на виживання найбільше впливала вартість квитка (тобто ті, хто заплатив більше, з більшою вірогідністю виживали). При цьому існує кореляція також між віком та ціною квитка (тобто старші люди платили більше). Одже з цього можна зробити висновок, що з більшою вірогідністю мали виживати старші люди. Проте візуалізація свідчить про те, що з більшою вірогідністю виживали молодші люди. Це може свідчити про недостовірність даних.
